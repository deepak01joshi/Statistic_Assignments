### 1. What is hypothesis testing in statistics?
Hypothesis testing is a statistical method used to make decisions about a population parameter based on sample data. It involves making an assumption (the null hypothesis) and determining the likelihood that this assumption is true given the data.

### 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- **Null Hypothesis (H0)**: A statement that there is no effect or no difference. It is the default or initial assumption.
- **Alternative Hypothesis (H1 or Ha)**: A statement that contradicts the null hypothesis; it represents a new claim or effect.



### 3. What is the significance level in hypothesis testing, and why is it important?
The significance level (denoted by alpha, α) is the threshold for rejecting the null hypothesis. Common values are 0.05, 0.01. It indicates the probability of making a Type I error.



### 4. What does a P-value represent in hypothesis testing?
The P-value is the probability of obtaining a test statistic at least as extreme as the one observed, assuming the null hypothesis is true.


### 5. How do you interpret the P-value in hypothesis testing?
- If **P-value < α**, reject H0.
- If **P-value ≥ α**, fail to reject H0.

### 6. What are Type 1 and Type 2 errors in hypothesis testing?
- **Type I Error**: Rejecting H0 when it is true.
- **Type II Error**: Failing to reject H0 when it is false.



### 7. What is the difference between a one-tailed and a two-tailed test?
- **One-tailed**: Tests for a deviation in one direction.
- **Two-tailed**: Tests for deviation in either direction.

### 8. What is the Z-test, and when is it used?
The Z-test is used when the population standard deviation is known, and the sample size is large (n > 30).

### 9. How do you calculate the Z-score?
Z = (X̄ - μ) / (σ / √n)
Where X̄ = sample mean, μ = population mean, σ = population std dev, n = sample size

### 10. What is the T-distribution, and when should it be used?
T-distribution is used when the population standard deviation is unknown and the sample size is small (n < 30).

### 11. What is the difference between a Z-test and a T-test?
- **Z-test**: Known population std dev.
- **T-test**: Unknown population std dev.

### 12. What is the T-test?
Used to determine if there is a significant difference between means when std dev is unknown.

### 13. Relationship between Z-test and T-test?
Both compare means; Z is for known std dev, T for unknown.


### 14. What is a confidence interval?
A range around a sample estimate that likely includes the population parameter.

### 15. What is the margin of error?
It measures the maximum expected difference between the true population parameter and a sample estimate.

### 16. How is Bayes' Theorem used in statistics?
Bayes' theorem calculates conditional probabilities and updates beliefs with new evidence.

### 17. What is the Chi-square distribution?
A distribution used in tests of independence and goodness of fit for categorical data.

### 18. Chi-square goodness of fit test?
It compares observed frequencies to expected frequencies.

### 19. What is the F-distribution?
Used to compare two variances or in ANOVA.

### 20. What is an ANOVA test?
ANOVA (Analysis of Variance) compares means across multiple groups.

### 21. Types of ANOVA tests:
- One-way ANOVA
- Two-way ANOVA


### 22. What is the F-test?
A test that compares two variances.

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
import numpy as np
from scipy import stats

# Known population mean and std dev
pop_mean = 100
pop_std = 15

# Simulate sample
np.random.seed(0)
sample = np.random.normal(loc=105, scale=15, size=50)
sample_mean = np.mean(sample)

# Z-test
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(len(sample)))
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"Z = {z:.2f}")
print(f"P-value = {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
#2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.
data = np.random.normal(100, 15, 100)
t_stat, p_val = stats.ttest_1samp(data, 100)
print(f"T = {t_stat:.2f}, P-value = {p_val:.4f}")
```

```python
#3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
def one_sample_z_test(sample, pop_mean, pop_std):
    sample_mean = np.mean(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(len(sample)))
    p_val = 2 * (1 - stats.norm.cdf(abs(z_score)))
    return z_score, p_val

z, p = one_sample_z_test(sample, 100, 15)
print(f"Z = {z:.2f}, P = {p:.4f}")
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(x < -1.96) | (x > 1.96), color='red', alpha=0.3, label='Rejection region')
plt.axvline(z, color='blue', linestyle='--', label='Z-score')
plt.title('Two-tailed Z-test Rejection Region')
plt.xlabel('Z')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
def visualize_type1_type2():
    x = np.linspace(-4, 4, 1000)
    y0 = stats.norm.pdf(x, 0, 1)   # H0
    y1 = stats.norm.pdf(x, 1, 1)   # H1 shifted

    plt.plot(x, y0, label='H0')
    plt.plot(x, y1, label='H1')
    plt.fill_between(x, y0, where=(x > 1.64), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, y1, where=(x < 1.64), color='blue', alpha=0.3, label='Type II Error')
    plt.legend()
    plt.title("Type I and II Errors")
    plt.grid(True)
    plt.show()

visualize_type1_type2()
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)
    margin = sem * stats.t.ppf((1 + confidence) / 2., len(data)-1)
    return mean - margin, mean + margin

ci = confidence_interval(sample)
print(f"95% Confidence Interval: {ci}")
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy import stats

def margin_of_error(data, confidence=0.95):
    sem = stats.sem(data)
    margin = sem * stats.t.ppf((1 + confidence) / 2., len(data) - 1)
    return margin

sample = np.random.normal(100, 15, 40)
moe = margin_of_error(sample)
print(f"Margin of Error: {moe:.2f}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
# P(A), P(B|A), P(B|¬A)
P_A = 0.01
P_B_given_A = 0.9
P_B_given_not_A = 0.05

# Bayes' Theorem
P_not_A = 1 - P_A
P_B = (P_B_given_A * P_A) + (P_B_given_not_A * P_not_A)
P_A_given_B = (P_B_given_A * P_A) / P_B

print(f"P(A|B) = {P_A_given_B:.4f}")
```

```python
#16. Implement an F-test using Python to compare the variances of two random samples.
group1 = np.random.normal(100, 10, 30)
group2 = np.random.normal(100, 15, 30)

f_stat = np.var(group1, ddof=1) / np.var(group2, ddof=1)
dfn = len(group1) - 1
dfd = len(group2) - 1
p_val = 1 - stats.f.cdf(f_stat, dfn, dfd)

print(f"F-test: F = {f_stat:.2f}, p = {p_val:.4f}")
```

```python
#24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
success_a, n_a = 45, 100
success_b, n_b = 55, 100

p1 = success_a / n_a
p2 = success_b / n_b
p_pool = (success_a + success_b) / (n_a + n_b)
se = np.sqrt(p_pool * (1 - p_pool) * (1/n_a + 1/n_b))
z = (p1 - p2) / se
p_val = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"Z for proportions: Z = {z:.2f}, p = {p_val:.4f}")
```

```python
#26. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
import numpy as np
from scipy.stats import chisquare

# Simulate observed frequencies
np.random.seed(42)
observed = np.random.randint(15, 35, size=5)

# Define expected frequencies (uniform distribution)
# Adjust expected frequencies so their sum equals the sum of observed frequencies
total_observed = np.sum(observed)
num_categories = len(observed)
expected = np.array([total_observed / num_categories] * num_categories)


# Perform the chi-square goodness-of-fit test
chi2_stat, p_val = chisquare(f_obs=observed, f_exp=expected)

print("Observed frequencies:", observed)
print("Expected frequencies:", expected)
print(f"Chi-square statistic = {chi2_stat:.2f}")
print(f"P-value = {p_val:.4f}")

# Interpretation
if p_val < 0.05:
    print("Reject the null hypothesis: The observed distribution is significantly different from the expected.")
else:
    print("Fail to reject the null hypothesis: The observed distribution fits the expected distribution well.")
```
